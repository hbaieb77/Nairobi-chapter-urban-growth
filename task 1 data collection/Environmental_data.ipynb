```python
# Helper functions
# This function was taken from here: https://www.geeksforgeeks.org/python-convert-nested-dictionary-into-flattened-dictionary/
# To flatten the dict for creating a tidier pandas df
def flatten_dict(dd, separator ='_', prefix =''):
    return { prefix + separator + k if prefix else k : v
             for kk, vv in dd.items()
             for k, v in flatten_dict(vv, separator, kk).items()
             } if isinstance(dd, dict) else { prefix : dd }
```
