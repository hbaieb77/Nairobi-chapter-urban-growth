```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('merged_output_v1.csv')
```

```python
df
```

```python
nigeria = df[df['country']=='Nigeria']
```

```python
nigeria
```

```python
nigeria = nigeria.drop(columns=['land_area_(sq._km)','population_density_(no._per_sq._km)','real_gdp'])
```

```python
nigeria
```

```python
nigeria = nigeria.reset_index()
```

```python
nigeria
```

```python
nigeria = nigeria.drop(columns=['index','unnamed:_0'])
```

```python
nigeria
```

```python
plt.plot(nigeria['year'],nigeria['population'])
plt.xticks(range(1950, 2035,5),rotation=90)
plt.title("Population of Nigeria over time")
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
train = nigeria[(nigeria['year'] >= 1950) & (nigeria['year'] < 2023)]
test = nigeria[(nigeria['year'] >= 2023) & (nigeria['year'] <= 2035)]
train
```

```python
test
```

```python
scaling  = StandardScaler()
train = train.copy()
test = test.copy()
```

```python
train['population_growth'] = scaling.fit_transform(train[['population_growth']])
test['population_growth'] = scaling.fit_transform(test[['population_growth']])
```

```python
train['population_growth_rate'] = scaling.fit_transform(train[['population_growth_rate']])
train['year'] = scaling.fit_transform(train[['year']])
test['population_growth_rate'] = scaling.fit_transform(test[['population_growth_rate']])
test['year'] = scaling.fit_transform(test[['year']])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoding = OneHotEncoder(sparse=False)
```

```python
train['city'] = encoding.fit_transform(train[['city']])
train['country'] = encoding.fit_transform(train[['country']])
test['city'] = encoding.fit_transform(test[['city']])
test['country'] = encoding.fit_transform(test[['country']])
```

```python
train
```

```python
x = train[['population_growth','population_growth_rate','year','city','country']]
y = train[['population']]
```

```python
testx = test[['population_growth','population_growth_rate','year','city','country']]
testy = test[['population']]
y = scaling.fit_transform(y)
testy = scaling.fit_transform(testy)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
predict_test = reg.predict(testx)
predict_train = reg.predict(x)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mean_squared_error(testy,predict_test)
```

```python
mean_squared_error(y,predict_train)
```

```python
r2_score(testy,predict_test)
```

```python
r2_score(y,predict_train)
```

```python
predict_test
```

```python
scaling.inverse_transform(predict_test)
```

```python
test1 = nigeria[(nigeria['year'] >= 2023) & (nigeria['year'] <= 2035)]
```

```python
plt.plot(test1['year'],predict_test, color='blue',linestyle='dashed', label='Predicted')
plt.plot(test1['year'], testy, color='red', label='Actual')
plt.title("Population of Nigeria over time")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend()
plt.show()
```

```python
pd.DataFrame(list(zip(scaling.inverse_transform(testy), scaling.inverse_transform(predict_test))),columns=['Actual', 'predicted'])
```
