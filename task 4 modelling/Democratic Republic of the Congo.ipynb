```python
train = df[(df['year'] >= 1950) & (df['year'] < 2018)]
test = df[(df['year'] >= 2018) & (df['year'] <= 2035)]
```

```python
train
```

```python
test
```

```python
print(train.shape)
print(test.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
encoding  = LabelEncoder()
train['City'] = encoding.fit_transform(train[['City']])
test['City'] = encoding.fit_transform(test[['City']])
```

```python
x_train = train[['Latitude','Longitude','year','City']]
y_train = train[['population']]
x_test = test[['Latitude','Longitude','year','City']]
y_test = test[['population']]
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
pred_train = reg.predict(x_train)
pred_test = reg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
print(mean_squared_error(y_train,pred_train))
print(mean_squared_error(y_test,pred_test))
print(r2_score(y_train,pred_train))
print(r2_score(y_test,pred_test))
print(mean_absolute_error(y_train,pred_train))
print(mean_absolute_error(y_test,pred_test))
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_features = PolynomialFeatures(degree=9)
X_train_poly = poly_features.fit_transform(x_train)
X_test_poly = poly_features.transform(x_test)
reg.fit(X_train_poly, y_train)

train_predictions = reg.predict(X_train_poly)
test_predictions = reg.predict(X_test_poly)
```

```python
print(r2_score(y_train,train_predictions))
print(r2_score(y_test,test_predictions))
```
