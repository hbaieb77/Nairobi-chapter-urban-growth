```python
import tensorflow as tf
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train,y_train = scaler.fit_transform(x_train),scaler.fit_transform(y_train)
x_test,y_test = scaler.fit_transform(x_test),scaler.fit_transform(y_test)
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
x_test.shape
```

```python
AUTO = tf.data.experimental.AUTOTUNE
train = tf.data.Dataset.from_tensor_slices((x_train,y_train)).batch(32).prefetch(AUTO)
test = tf.data.Dataset.from_tensor_slices((x_test,y_test)).batch(32).prefetch(AUTO)
```

```python
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SpatialDropout1D, Dense, Dropout, LSTM,Reshape
```

```python
n_lstm = 128
model1 = Sequential()
model1.add(Dense(256,input_shape=[5],activation='relu'))
model1.add(Dropout(0.2))
model1.add(Reshape((1, 256))) 
model1.add(LSTM(n_lstm, return_sequences=False))
model1.add(Dropout(0.2))
model1.add(Dense(1, activation='linear'))
```

```python
model1.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_absolute_error'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
history = model1.fit(train,epochs=1000, batch_size=64,validation_data=(test),callbacks =[early_stop])
```

```python
res=model1.predict(test)
predected_res = pd.DataFrame(res)
actual_res = pd.DataFrame(y_test)
det = pd.concat([actual_res, predected_res], join = 'outer', axis = 1)
print(det)
```
