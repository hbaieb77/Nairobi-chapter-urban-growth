### Handling Missing Values

```python
import numpy as np
df = df.replace(' ', np.nan)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df_unpivoted = pd.melt(df, id_vars=["Country", "Indicator"], var_name="Year", value_name="Value")
```

```python
df_unpivoted
```

```python
df_pivoted = df_unpivoted.pivot_table(index=["Country", "Year"], columns="Indicator", values="Value")
```

```python
df_pivoted
```

```python
df_pivoted = df_pivoted.reset_index()
```

```python
df_pivoted
```

```python
df_pivoted.isna().mean()
```

```python
df_pivoted = df_pivoted.drop(columns=['Average Household Size'], axis=1)
```

```python
df_pivoted.isna().sum()
```
