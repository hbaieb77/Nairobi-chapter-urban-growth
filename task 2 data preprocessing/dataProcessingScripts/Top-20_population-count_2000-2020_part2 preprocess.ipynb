```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
missing_data = df.isna()
imputer = IterativeImputer()
```

```python
imputed_data = imputer.fit_transform(df[['Adjusted-Population-Count_2000','Adjusted-Population-Count_2005','Adjusted-Population-Count_2010','Adjusted-Population-Count_2015','Adjusted-Population-Count_2020']])
```

```python
imputed_data_df = pd.DataFrame(imputed_data, columns=df.columns[2:7])
imputed_missing_data = imputed_data_df.isna()
imputed_missing_data.sum()
```

```python
imputed_data_df
```

```python
exclude_cols = ['Adjusted-Population-Count_2000', 'Adjusted-Population-Count_2005', 'Adjusted-Population-Count_2010', 'Adjusted-Population-Count_2015', 'Adjusted-Population-Count_2020']
df = df.loc[:, ~df.columns.isin(exclude_cols)]
```

```python
df
```

```python
df = pd.concat([df,imputed_data_df],axis=1)
```
