```python
import pandas as pd 
```

```python
df = pd.read_csv("../../src/data/city_growth_models/country_agriculture_land_use_1970-2022.csv")
```

```python
df.head(20) 
```

```python
df = df.drop(['Unnamed: 0', 'Category', 'Unnamed: 57', 'Unit'], axis=1)
```

```python
df.head(10)
```

```python
df = df.melt(id_vars=['Country', 'Indicator'], var_name='Year', value_name='Value')
```

```python
df.head(5)
```

```python
df = df.pivot(index=['Country', 'Year'], columns='Indicator', values='Value')
```

```python
df = df.reset_index()
```

```python
df.head()
```

```python
df.to_csv('../../task 2 data preprocessing/processedData/country_agriculture_land_use_1970-2022_processed.csv',index=False)
```
