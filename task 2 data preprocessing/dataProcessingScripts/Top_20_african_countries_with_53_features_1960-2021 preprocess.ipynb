```python
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
sns.set_style("whitegrid")
```

```python
sns.displot(df['GDP (constant 2015 US$) [NY.GDP.MKTP.KD]'])
plt.show()
sns.displot(df['GDP growth (annual %) [NY.GDP.MKTP.KD.ZG]'])
plt.show()
sns.displot(df['GDP per capita (constant 2015 US$) [NY.GDP.PCAP.KD]'])
plt.show()
sns.displot(df['GDP per capita growth (annual %) [NY.GDP.PCAP.KD.ZG]'])
plt.show()
sns.displot(df['Population density (people per sq. km of land area) [EN.POP.DNST]'])
plt.show()
sns.displot(df['Population growth (annual %) [SP.POP.GROW]'])
plt.show()
sns.displot(df['Urban population growth (annual %) [SP.URB.GROW]'])
plt.show()
```

```python
from sklearn.impute import SimpleImputer
imputer_median = SimpleImputer(strategy='median')
df['GDP (constant 2015 US$) [NY.GDP.MKTP.KD]'] = imputer_median.fit_transform(df[['GDP (constant 2015 US$) [NY.GDP.MKTP.KD]']])
df['GDP per capita (constant 2015 US$) [NY.GDP.PCAP.KD]'] = imputer_median.fit_transform(df[['GDP per capita (constant 2015 US$) [NY.GDP.PCAP.KD]']])
df['Population density (people per sq. km of land area) [EN.POP.DNST]'] = imputer_median.fit_transform(df[['Population density (people per sq. km of land area) [EN.POP.DNST]']])


```

```python
imputer_mean = SimpleImputer(strategy='mean')
df['GDP growth (annual %) [NY.GDP.MKTP.KD.ZG]'] = imputer_mean.fit_transform(df[['GDP growth (annual %) [NY.GDP.MKTP.KD.ZG]']])
df['GDP per capita growth (annual %) [NY.GDP.PCAP.KD.ZG]'] = imputer_mean.fit_transform(df[['GDP per capita growth (annual %) [NY.GDP.PCAP.KD.ZG]']])
df['Population growth (annual %) [SP.POP.GROW]'] = imputer_mean.fit_transform(df[['Population growth (annual %) [SP.POP.GROW]']])
df['Urban population growth (annual %) [SP.URB.GROW]'] = imputer_mean.fit_transform(df[['Urban population growth (annual %) [SP.URB.GROW]']])
```
