```python
from scipy.stats import chi2_contingency

# Create missingness indicator variable
df['NA'] = np.where(df['service_charge'].isnull(), 1, 0)
# Test for MCAR using chi-squared test of independence
chi2, p, dof, expected = chi2_contingency(pd.crosstab(df['NA'], df.drop('NA', axis=1).isnull().any(axis=1)))
print('Chi-squared test p-value:', p)
```
