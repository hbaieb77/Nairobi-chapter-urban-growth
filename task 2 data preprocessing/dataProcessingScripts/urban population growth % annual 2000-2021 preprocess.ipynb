```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import scipy.stats as stats
warnings.filterwarnings('ignore') 
sns.set_style('whitegrid') 
```

```python
df = pd.read_csv('../../src/data/city_growth_models/urban population growth % annual 2000-2021.csv')
```

```python
df.head() 
```

```python
df = df.drop(columns=['Country Code' ,'Indicator Code' ],axis=1)  
df.head()
```

```python
df.columns 
```

```python
print(df['Indicator Name'].unique()) 
```

```python
df_melted = pd.melt(df, id_vars=['Country Name', 'Indicator Name'], var_name='Year', value_name='Value')

# convert Year column to integer
df_melted['Year'] = df_melted['Year'].astype(int) 
```

```python
df_melted
```

```python
df_melted.rename(columns={'Value':'Urban population growth (annual %)'} , inplace=True )
df_melted.rename(columns={'Country Name':'Country'} , inplace=True )
df_melted.drop(columns=['Indicator Name'] , inplace=True ) 
```
