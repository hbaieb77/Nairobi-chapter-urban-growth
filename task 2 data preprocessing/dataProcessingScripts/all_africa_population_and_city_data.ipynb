```python
import numpy as np
import pandas as pd
import seaborn as sns
from plotly.subplots import make_subplots
import matplotlib.pyplot as plt

def preprocess(df ,country_name ) :

    df_Algeria_copy = df[[
                                'Name' , 'Population Census (C) 1977-02-12' ,
                                'Population Census (C) 1987-03-20' , 
                                'Population Census (C) 1998-08-28',
                                'Population Census (C) 2008-04-14'
                                ]].copy()
    
    df_Algeria_copy.rename(columns={         'Name' :'City Name' , 
                                'Population Census (C) 1977-02-12' : '1977' ,
                                'Population Census (C) 1987-03-20' : '1987', 
                                'Population Census (C) 1998-08-28' : '1998',
                                'Population Census (C) 2008-04-14' : '2008'
                                } , inplace=True )
    
    df_Algeria_melted = pd.melt(df_Algeria_copy, id_vars=['City Name'], var_name='Year', value_name='Value')
    df_Algeria_melted['Year'] = df_Algeria_melted['Year'].astype(int) 
    df_Algeria_melted['Country'] = 'Algeria'
    return df_Algeria_melted  
```
