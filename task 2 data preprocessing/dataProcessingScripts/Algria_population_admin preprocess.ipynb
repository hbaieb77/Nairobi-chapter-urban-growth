```python
import pandas as pd
```

```python
df = pd.read_csv('Algria_population_admin.csv')
```

```python
df
```

```python
df = df.drop(columns=['Unnamed: 0','5'],axis=1)
```

```python
df
```

```python
columns = df.iloc[0]
df = df.iloc[1:]
```

```python
df.columns = columns
```

```python
df
```

```python
df.info()
```

```python
df = df.rename(columns={'Name':'City'})
```

```python
df
```

```python
df['Population Census 1998-06-25'] = df['Population Census 1998-06-25'].str.replace(',','')
```

```python
df['Population Census 2008-04-14'] = df['Population Census 2008-04-14'].str.replace(',','')
```

```python
df
```

```python
df['Population Census 1998-06-25'] = df['Population Census 1998-06-25'].astype('int')
```

```python
df['Population Census 2008-04-14'] = df['Population Census 2008-04-14'].astype('int')
```

```python
df
```

```python
df.describe()
```

```python
melted_df = df.melt(id_vars=['City','Status','Native'],var_name='Year',value_name='Value')
```

```python
melted_df
```

```python
melted_df['Year'] = melted_df['Year'].replace('Population Census 1998-06-25', '1998')
melted_df['Year'] = melted_df['Year'].replace('Population Census 2008-04-14', '2008')
```
