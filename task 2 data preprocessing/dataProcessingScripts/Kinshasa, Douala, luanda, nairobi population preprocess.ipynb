### Handle Missing Values

```python
median = df.dropna().median()
df = df.fillna(median)
```

```python
median1 = df1.dropna().median()
df1 = df1.fillna(median1)
```

```python
median2 = df2.dropna().median()
df2 = df2.fillna(median2)
```

```python
median3 = df3.dropna().median()
df3 = df3.fillna(median3)
```

```python
print(df.isnull().sum())
print(df1.isnull().sum())
print(df2.isnull().sum())
print(df3.isnull().sum())
```
