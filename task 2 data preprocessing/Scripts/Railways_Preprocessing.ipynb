```python
# Split the Country Name 
split_df = df['Country Name'].str.split(r'\.|, ', expand=True)
```

```python
split_df.head(5)
```

```python
# Ignore warning
df.loc[:, 'Country Code'] = split_df[0]
df.loc[:, 'Country'] = split_df[1]
df.loc[:, 'Code'] = split_df[2]
```

```python
# Drop the columns
df = df.drop(columns=['Country Name'])
```

```python
# New shape of df
df.shape
```

```python
# 
df.head(5)
```

```python
# Filter in country values only
# Will use country levels later
country_value_counts = df['Country'].value_counts()
# Print the value_counts
print("Value counts for the 'Country' column:")
print(country_value_counts)
```

```python
# Strip the whitespaces in Country column
df['Country'] = df['Country'].str.strip()
```

```python
# Remove all the extra spaces from the 'Country' column
df['Country'] = df['Country'].str.replace('\s+', ' ', regex=True).str.strip()
```

```python
# Filter out non-country columns 
# Create two df with filtered values and country levels
values_to_filter_out = ['Low Income', 'ECOWAS', 'SSA', 'State Owned Railways', 'Non- Stated Owned Railways', 'EAC', 'COMESA', 'Resource Rich', 'CEMAC', 'SADC', 'Middle Income']
```

```python
# Filtered df based on countries and levels
# Create a df filtered based on country
df_country = df[~df['Country'].isin(values_to_filter_out)]
```

```python
# Count the new values for df_country
df_countrycount = df_country['Country'].value_counts()
```

```python
print(df_countrycount)
```

```python
# Create a new pivot table based on Series value
pivoted_countrydf = df_country.pivot_table(index=['year', 'Country', 'Country Code'], columns='Series Name', values='Value')
```

```python
pivoted_countrydf = pivoted_countrydf.reset_index()
```

```python
# Display new dataframes rows and columns
pivoted_countrydf.head(5)
```

```python
pivoted_countrydf.tail(5)
```

```python
pivoted_countrydf.to_csv('African_RailwaysbyCountry.csv', index=False)
```

```python
# Rename the Country Column to Income CountryIncome level
railways_byeconomics = railways_byeconomics.rename(columns={'Country': 'Economic Status', 'Code': 'Railways Status'})
```

```python
# Drop Country Code
railways_byeconomics.drop(columns=['Country Code'], inplace=True)
```

```python
# Final columns
railways_byeconomics.columns
```

```python
# Create a pivot table to create railways information based on country income levels
railways_byeconomics = railways_byeconomics.pivot_table(index=['year', 'Economic Status', 'Railways Status'], columns='Series Name', values='Value')
```

```python
# Reset the index
railways_byeconomics = railways_byeconomics.reset_index()
```
