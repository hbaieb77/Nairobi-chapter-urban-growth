```python
# Save a copy of the old df
df_copy = df.copy()
```

```python
# Drop Unnecessary columns 
df_copy.drop(columns= ['Country Code','Series Name','Series Code'],inplace=True, axis=1)
```

```python
df_copy.head(5)
```

```python
railways_data = df_copy.melt(id_vars=['Country Name'], var_name='year', value_name='Value')
```

```python
railways_data
```

```python
# Split the dataset
year = railways_data['year'].str.split('\[', expand=True)[0].str.strip()
```

```python
# Replace the string with correct year
year_map = {'LAV': '2000', 'MRV': '1950', 'MRV1': '2006', 'MRV2': '2008'}

def replace_year(x):
    if isinstance(x, str):
        for key, value in year_map.items():
            if key in x:
                return value
        return x
    else:
        year = int(x[-4:])
        if 1990 <= year <= 2009:
            return x
        else:
            return np.nan

year = year.apply(lambda x: replace_year(x))
```

```python
year
```

```python
# Replace the new column
railways_data['year'] = year
```

```python
railways_data.head(5)
```

```python
railways_data.isnull().sum()
```

```python
# Drop all missing values
railways_data.dropna(how='any',inplace=True)
```

```python
railways_data
```

```python
# Count empty strings
railways_data['Value'].value_counts()['..']
```

```python
railways_data['Value'] = railways_data['Value'].replace('..','NAN').astype(float)
```

```python
railways_data = railways_data.dropna()
```

```python
# Final shape of railways_data
railways_data.shape[0], railways_data.shape[1]
```

```python
# Drop columns with missing values rows
railways_data = railways_data.dropna(how='all', inplace=True)
```

```python
# Replace all empty strings with " " with NAN values
railways_data['Value'] =  railways_data['Value'].replace('',np.nan)
```

```python
# Conver the values to numeric 
railways_data['Value'] = pd.to_numeric(railways_data['Value'], errors='coerce')
```
